# Limit

```python
from sympy import *
from sympy.plotting import plot
import warnings; warnings.simplefilter('ignore')
```

```python
x, y, z = symbols('x y z')
```

```python
f1 = x**(1/x)
f1
```

```python
p1 = plot(f1, show=True)
```

```python
limit(f1, x, oo)
```

```python
f2 = x**x
f2
```

```python
p2 = plot(f2, show=True)
```

```python
limit(f2, x, 0)
```

```python
f3 = x**(x**x)-x**x
f3
```

```python
p3 = plot(f3, show=True)
```

```python
limit(f3, x, 0)
```

```python
f4=(x**2*sin(1/x) + x)/((1+x)**(1/x)-exp(1))
f4
```

```python
plot(f4, show=True)
```

```python
limit(f4, x, 0)
```

```python
limit(ln(x**x/factorial(x))/x, x, oo)
```

```python
limit(x**2/(2*x+1)*sin(pi/x), x, oo)
```
